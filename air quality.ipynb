```python
import io                    # Input/Output
import zipfile               # Help read file from Zip file without decompressing files
from pathlib import Path     # Objective filesystem paths management
import numpy as np           # Numpy array algebra and statistics
import pandas as pd          # Pandas dataframe

# Notebook magics to automatically reload libraries
%load_ext autoreload
%autoreload 2
```

```python
# Load the data
cdir = Path("./")      # Current directory

with zipfile.ZipFile(cdir / "india_air_quality_data (1).zip", mode="r") as zipobj:
    # File name
    zipinfos = [info for info in zipobj.infolist() if info.filename.endswith("csv")]
    print(zipinfos[0].filename)

    # Read data
    india_aq = pd.read_csv(zipobj.open(zipinfos[0].filename), encoding='iso8859-1')
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Time series plot of pollutant levels
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='pm2_5', data=india_aq)
plt.title('PM2.5 Levels Over Time')
plt.xlabel('Date')
plt.ylabel('PM2.5 Level')
plt.show()

# Box plot of pollutant levels by location
plt.figure(figsize=(12, 8))
sns.boxplot(x='location', y='pm2_5', data=india_aq)
plt.title('PM2.5 Levels by Location')
plt.xlabel('Location')
plt.ylabel('PM2.5 Level')
plt.xticks(rotation=45)
plt.show()

# Histogram of pollutant levels
plt.figure(figsize=(12, 8))
sns.histplot(data=india_aq[['pm2_5', 'so2', 'no2']], bins=20, kde=True)
plt.title('Distribution of Pollutant Levels')
plt.xlabel('Pollutant Level')
plt.ylabel('Frequency')
plt.legend(['PM2.5', 'SO2', 'NO2'])
plt.show()

# Correlation matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(india_aq[['pm2_5', 'so2', 'no2']].corr(), annot=True, cmap='plasma')
plt.title('Correlation Matrix of Pollutants')
plt.show()

# Bar plot of pollutant levels by location
plt.figure(figsize=(12, 8))
sns.barplot(x='location', y='pm2_5', data=india_aq)
plt.title('Average PM2.5 Levels by Location')
plt.xlabel('Location')
plt.ylabel('Average PM2.5 Level')
plt.xticks(rotation=45)
plt.show()
 #Bar plot for number of observations by state
plt.figure(figsize=(12, 8))
sns.countplot(y='state', data=india_aq, order=india_aq['state'].value_counts().index)
plt.title('Number of Observations by State')
plt.xlabel('Number of Observations')
plt.ylabel('State')
plt.show()
```
